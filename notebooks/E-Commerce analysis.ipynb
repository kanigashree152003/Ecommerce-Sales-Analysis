```python
reference_date = df['Date'].max()
reference_date
```

```python
rfm = df.groupby('Customer_Name').agg({
    'Date': 'max',
    'Transaction_ID': 'count',
    'Total_Cost': 'sum'
}).reset_index()

rfm.columns = ['Customer_Name', 'Last_Purchase_Date', 'Frequency', 'Monetary']
```

```python
rfm['Recency'] = (reference_date - rfm['Last_Purchase_Date']).dt.days
rfm.head()
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
```

```python
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
```

```python
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

```python
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

rfm.head()
```

```python
def segment_customer(row):
    if row['R_Score'] == 5 and row['F_Score'] == 5 and row['M_Score'] == 5:
        return 'High Value Customer'
    elif row['R_Score'] >= 4 and row['F_Score'] >= 4:
        return 'Loyal Customer'
    elif row['R_Score'] >= 4:
        return 'Potential Loyalist'
    elif row['R_Score'] <= 2:
        return 'At Risk Customer'
    else:
        return 'Regular Customer'

rfm['Customer_Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.arima.model import ARIMA
```

```python
data = pd.read_csv("time_series_sales_dataset.csv")
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
```

```python
data.head()
```

```python
train_size = int(len(data) * 0.8)
train = data[:train_size]
test = data[train_size:]
```

```python
print(train.shape)
print(test.shape)
```

```python
model = ARIMA(train['Total_Cost'], order=(1,1,1))
model_fit = model.fit()
```

```python
predictions = model_fit.forecast(steps=len(test))
```

```python
rmse = sqrt(mean_squared_error(test['Total_Cost'], predictions))
print("RMSE:", rmse)
```

```python
plt.figure()
plt.plot(test.index, test['Total_Cost'], label='Actual Sales')
plt.plot(test.index, predictions, label='Predicted Sales')
plt.legend()
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
future_forecast = model_fit.forecast(steps=180)
```

```python
plt.figure()
plt.plot(data.index, data['Total_Cost'], label='Historical Sales')
plt.plot(future_forecast.index, future_forecast, label='Future Forecast')
plt.legend()
plt.title("Sales Forecast")
plt.show()
```
